```python
import numpy as np
import pandas as pd
from pandas.tseries.offsets import *
```

```python
#1  add column to existing dataframe and calculate values
df = pd.DataFrame({'X':[7,2,0,3,4,2,5,0,3,4]})  #create dataframe
df.insert(1,'Y', 0)  #add new column with zeros for values
count1 = 0
#create loop to iterate through all the existing values to use as basis for new values
for i in range(len(df)):
    if df['X'][i] == 0:
        count1 = 0
        df['Y'][i] = count1
    else:
        count1 += 1
        df['Y'][i] = count1
df

```

```python
#2 all 2015 business days being used as index for random numbers
dates = pd.DataFrame()
dates = pd.DatetimeIndex(start='1/1/2015', end='12/31/2015',freq=BDay())
random_ = np.random.randint(1,1000,len(dates))
weeks = pd.DataFrame(data=random_, index=dates)
weeks
```

```python
#3 Find the sums for every Wednesday 
SumofWed = 0
#find dates with dayofweek=2, take coresponding value of business day series and sum value
for day in dates:
    if day.dayofweek == 2:
        SumofWed += weeks.loc[day,0]
            
SumofWed
```

```python
#4 Find Average for each month
#loop to get month values 01-12
for month_ in range(1,13):
    month_ = "{0:0=2d}".format(month_)  #format range into 2 digits
    temp = weeks['2015-'+str(month_)]  #get a dataframe of each month
    print("The mean for %s is %d" % (month_, temp[0].mean()))   # calculate mean of each month
```

```python
#5 find max value in 4 month blocks
# create dataframe for each month
df1 = weeks['2015-01']
df2 = weeks['2015-02']
df3 = weeks['2015-03']
df4 = weeks['2015-04']
df5 = weeks['2015-05']
df6 = weeks['2015-06']
df7 = weeks['2015-07']
df8 = weeks['2015-08']
df9 = weeks['2015-09']
df10 = weeks['2015-10']
df11 = weeks['2015-11']
df12 = weeks['2015-12']
# concatenate into 4 month groups
group1 = pd.concat([df1, df2, df3, df4], axis=0)
group2 = pd.concat([df5, df6, df7, df8], axis=0)
group3 = pd.concat([df9, df10, df11, df12], axis=0)

print('Week with highest value in first four months')
print(group1.loc[group1[0].idxmax()])
print()
print('Week with highest value in second four months')
print(group2.loc[group2[0].idxmax()])
print()
print('Week with highest value in third four months')
print(group3.loc[group3[0].idxmax()])
print()

```
